```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
Boston = pd.read_csv('../../ALL_CSV_FILES/Boston.csv')
```

```python
Boston
```

```python
Boston = Boston.drop(['col'], axis='columns')
```

```python
Boston.shape
```

```python
Boston['col2'] = np.zeros(Boston.shape[0])
Boston.head()
```

```python
Boston.shape
```

```python
fig, axes = plt.subplots(5, 3, figsize=(12, 12))

boston_col = np.array(Boston.columns).reshape(5, 3)
ynumber = np.random.choice([0,1,2])

for i in range(5):
    for j in range(3):
        y_col = np.random.choice([0,1,2])
        axes[i,j].scatter(Boston[boston_col[i][j]], Boston[boston_col[0][y_col]], marker='o');
        axes[i,j].set_title(f"{boston_col[i][j]} vs. {boston_col[0][y_col]}")
        axes[i,j].set_xlabel(boston_col[i][j])
        axes[i,j].set_ylabel(boston_col[0][y_col])
```

```python
fig, ax = plt.subplots(figsize=(8,8))
ax.scatter(Boston['zn'], Boston['crim'])
ax.set_title('ZN vs. Crime Rate')
ax.set_xlabel('ZN')
ax.set_ylabel('Crime Rate')
```

```python
fig, ax = plt.subplots(figsize=(8,8))
ax.scatter(Boston['indus'], Boston['crim'], color='green')
ax.set_title('Industry vs. Crime Rate')
ax.set_xlabel('Industry')
ax.set_ylabel('Crime Rate');
```
